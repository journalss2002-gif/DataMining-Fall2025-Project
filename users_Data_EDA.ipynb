(ChatGPT 5.2, 2025) Prompt: I want to visualize all the numeric features in a histogram, give me a for loop that goes through each feature and visualize it.

```python
#selecting numerical columns 
continuous_cols = df.select_dtypes(include=['float64', 'int64']).columns
#removing 'id' column from the list of continuous columns
numeric_cols_no_id = [c for c in continuous_cols if c != 'id']

#visualizing the data using histograms for continuous variables
for col in numeric_cols_no_id:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], bins=30, edgecolor='black')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

(ChatGPT 5.2, 2025) Prompt: I want to visualize all the numeric features in a boxplot, give me a for loop that goes through each feature and visualize it.

```python
#boxplots for numerical features
for col in numeric_cols_no_id:
    plt.figure(figsize=(6, 4))
    plt.boxplot(
        df[col],
        vert=False,          
        patch_artist=True,
        boxprops=dict(linewidth=1),
        medianprops=dict(linewidth=2),
        whiskerprops=dict(linewidth=1),
        capprops=dict(linewidth=1)
    )
    plt.title(f"Box Plot of {col}")
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
#visualizing the data using bar plots for categorical variables
if "gender" in df.columns:
    print("Frequency table for gender:")
    print(df["gender"].value_counts())
    print("-" * 60)

    plt.figure(figsize=(6, 4))
    df["gender"].value_counts().plot(kind="bar", color=["skyblue", "salmon"])
    plt.title("Gender Distribution")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

(ChatGPT 5.2, 2025) Prompt: I want to pick some of the highly correlated features and make specific scatter plots for two features. I want to do it by using a for loop so I dont do each individually. How can I do that?

```python
#chose pairs based on the correlation above 
scatter_pairs = [
    ('current_age', 'num_credit_cards'),
    ('credit_score', 'num_credit_cards')
]

for x, y in scatter_pairs:
    if x in df.columns and y in df.columns:
        plt.figure(figsize=(6, 4))
        sns.scatterplot(data=df, x=x, y=y, alpha=0.7)
        plt.title(f"{y} vs {x}")
        plt.tight_layout()
        plt.show()
```
