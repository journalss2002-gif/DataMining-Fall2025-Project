```python
#imporing needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
df = pd.read_csv('users_data.csv')
df.head(10)
```

```python
#the number of rows and columns 
print("Shape:", df.shape)

#the data type 
print("Data Types:\n", df.dtypes)
```

```python
# getting more information about the dataset
df.info()
```

```python
# getting the duplicate rows in the data, according to the results there are none 
df.duplicated().sum()
```

```python
# checking the number of missing values in each column 
df.isnull().sum()
```

```python
#summary for numerical columns 
print("Numerical Summary:\n", df.describe())
```

```python
#summary including categorical columns
print("Full Summary:\n", df.describe(include='all'))
```

```python
continuous_cols = df.select_dtypes(include=['float64', 'int64']).columns
numeric_cols_no_id = [c for c in continuous_cols if c != 'id']
#visualizing the data usng histograms for continuous variables
for col in numeric_cols_no_id:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
#boxplots for numerical features
plt.figure(figsize=(15,10))
df[numeric_cols_no_id].boxplot(rot=90)
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
#visualizing the data using bar plots for categorical variables
if "gender" in df.columns:
    print("Frequency table for gender:")
    print(df["gender"].value_counts())
    print("-" * 60)

    plt.figure(figsize=(6, 4))
    df["gender"].value_counts().plot(kind="bar", color=["skyblue", "salmon"])
    plt.title("Gender Distribution")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# getting the correlation of the numeric values 
df.corr(numeric_only = True)
```

```python
# visualizing the correlated features
# https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html

correlated_features = df.corr(numeric_only = True)
plt.figure(figsize=(8,6))
sns.heatmap(df[continuous_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Numerical Features")
plt.show()
```

```python
#chose pairs based on the heatmap abpve 
scatter_pairs = [
    ('current_age', 'num_credit_cards'),
    ('retirement_age', 'credit_score'),
     ('retirement_age', 'num_credit_cards'),
    ('credit_score', 'num_credit_cards')
]

for x, y in scatter_pairs:
    if x in df.columns and y in df.columns:
        plt.figure(figsize=(6, 4))
        sns.scatterplot(data=df, x=x, y=y, alpha=0.7)
        plt.title(f"{y} vs {x}")
        plt.tight_layout()
        plt.show()
```
